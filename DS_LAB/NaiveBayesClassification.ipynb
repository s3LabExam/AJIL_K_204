```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
csv=pd.read_csv('/content/tennis.csv')
```

```python
csv.head()
```

```python
x=csv.drop('play',axis=1)
```

```python
x
```

```python
y=csv['play']
```

```python
y
```

```python
windy=[]
outlook=[]
temp=[]
humidity=[]
for i in x['windy']:
    if i==False:
      windy.append(0)
    else:
      windy.append(1)
for i in x['humidity']:
  if i=='high':
    humidity.append(1)
  else:
    humidity.append(0)
for i in x['outlook']:
  if i=="rainy":
    outlook.append(0)
  elif i=="sunny":
    outlook.append(1)
  else:
    outlook.append(2)
for i in x['temp']:
  if i=="cool":
    temp.append(0)
  elif i=="mild":
    temp.append(1)
  else:
    temp.append(2)
```

```python
X=[]
for i in range(0,len(outlook)):
    X.append([outlook[i],temp[i],humidity[i],windy[i]])
```

```python
X
```

```python
Y=[]
for i in y:
  if i=="no":
    Y.append(0)
  else:
    Y.append(1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
x_train
```

```python
y_train
```

```python
n=GaussianNB()
```

```python
n.fit(x_train,y_train)
```

```python
p=n.predict(x_test)
```

```python
confusion_matrix(y_test,p)
```

```python
accuracy_score(y_test,p)
```

```python
classification_report(y_test,p)
```
